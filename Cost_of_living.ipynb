```python
from geopy.geocoders import ArcGIS
```

```python
geolocator = ArcGIS()

def get_coordinates(city):
    try:
        location = geolocator.geocode(city)
        if location:
            return location.latitude, location.longitude
        else:
            return 0, 0
    except Exception as e:
        print("oh, shit!!!!! :", e)
        return 0, 0
```

```python
df['Latitude'], df['Longitude'] = zip(*df['city'].map(get_coordinates))
```

```python
df=df[df['Latitude'] != 0]
```

```python
import pycountry_convert as pc
def country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except Exception as e:
        return None

df['continent'] = df['country'].apply(country_to_continent)
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/A-akis/A-akis.github.io/main/df-continents.csv')
```

```python
data=df
```

```python
# Convert data to DataFrame
df = pd.DataFrame(data)

# Calculate counts
continent_counts = df.groupby("continent").agg(country=("country", "nunique"), city=("city", "nunique")).reset_index()

# Create a list to store hierarchical data
tree_data = []

# Populate tree_data based on counts
for index, row in continent_counts.iterrows():
    continent = row["continent"]
    country_count = row["country"]
    city_count = row["city"]

    # Append country count
    tree_data.append({"continent": continent, "category": 0, "value": country_count})

    # Append city count
    tree_data.append({"continent": continent, "category": 1, "value": city_count})

# Create DataFrame from list
tree_data_df = pd.DataFrame(tree_data)

print(tree_data_df)
```

```python
tree_data_df.to_csv('radial-map.csv', index=False)
```
